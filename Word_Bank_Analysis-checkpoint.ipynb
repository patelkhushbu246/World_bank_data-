```python
# Average inflation across all countries
inflation_by_country = df.groupby('country_name')['Inflation (CPI %)'].mean()

# Sort top or bottom values
top_inflation = inflation_by_country.sort_values(ascending=False).head(20)
print(top_inflation)
```

```python
ax = top_inflation.plot(kind='bar', color='#78B9B5', edgecolor='black', figsize=(10, 6))
plt.ylabel('Inflation (CPI %)')
plt.xlabel('Countries')
plt.title('Top 20 Countries by Average Inflation')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()

# Add value labels
for i, val in enumerate(top_inflation):
    ax.text(i, val + val*0.01, f'{val:,.0f}', ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select indicators
features = ['Inflation (CPI %)', 'GDP Growth (% Annual)', 'Unemployment Rate (%)']
df_cluster = df.groupby('country_name')[features].mean()

# Scale data
scaler = StandardScaler()
scaled = scaler.fit_transform(df_cluster)

# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
df_cluster['Cluster'] = kmeans.fit_predict(scaled)

print(df_cluster.sort_values('Cluster'))

```

```python

df_cluster_reset = df_cluster.reset_index()
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_cluster_reset,
                x='GDP Growth (% Annual)',
                y='Unemployment Rate (%)',
                hue='Cluster',
                palette='Set2',
                style='Cluster',
                s=111)
for cluster_id in sorted(df_cluster['Cluster'].unique()):
    members = df_cluster[df_cluster['Cluster'] == cluster_id].index.tolist()
    print(f"Cluster {cluster_id} countries:\n", members[:10], "\n")

plt.title('Macroeconomic Clusters of Countries')
plt.xlabel('GDP Growth (% Annual)')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="color:yellow">We observe that countries in Cluster 1 have significantly higher unemployment rates compared to the other clusters. Cluster 3 countries exhibit low GDP growth, indicating weaker economic performance. In contrast, Cluster 0 countries show the highest GDP growth, likely supported by low unemployment rates. Although Cluster 2 also has low unemployment, its GDP growth remains relatively modest, suggesting other limiting factors might be affecting growth in those economies.   </span>

```python
from sklearn.linear_model import LinearRegression

# Select features and target
features = ['Unemployment Rate (%)', 'Interest Rate (Real, %)', 'Inflation (CPI %)']
#df_model = df.dropna(subset=features + ['GDP Growth (% Annual)'])
target = 'GDP Growth (% Annual)'
mask = df[features + [target]].all(axis=1)
X = df.loc[mask, features]
y = df.loc[mask, target]

# Train model
model = LinearRegression()
model.fit(X, y)

# Coefficients
for feat, coef in zip(features, model.coef_):
    print(f'{feat}: {coef:.2f}')
```

<span style="color:yellow"> In this macroeconomic forecasting model, unemployment emerges as the strongest predictor of economic slowdown. A 1% increase in unemployment is associated with a 0.11 unit decline in the forecasted outcome (e.g., GDP growth), indicating a significant inverse relationship. Real interest rates show a very mild positive relationship with the forecasted variable, suggesting that modest rate increases might align with periods of economic stability. Inflation has a slight negative impact, consistent with inflationary pressures dampening economic momentum. Overall, the results align with standard macroeconomic theory, emphasizing the critical role of employment levels in sustaining growth.</span>
